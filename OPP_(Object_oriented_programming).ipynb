```python
class Thermostat:
  def __init__(self, starting_temp, max_temp):
    self.current_temp = starting_temp
    self.max_temp = max_temp

  def set_temp(self, new_temp):

    if new_temp> self.max_temp:
      return print("Warning: This temp exceeds max_temp")
    else:
      self.current_temp = new_temp
      return print(f"The current temperature is {self.current_temp}")

  def get_farhenheit(self):
    farhenheit_temp = (self.current_temp*1.8)+32
    return farhenheit_temp



Engine_1 = Thermostat(100, 500)
Engine_2 = Thermostat(200, 800)

Engine_1.set_temp(600)
Engine_2.set_temp(700)

print(f"The farhenheit temperature of Engine_1 is {Engine_1.get_farhenheit(): 0.3f}")

```

**Inheritance Practice**



```python

#parent class Structural_Member
class Structural_Member:
  def __init__(self, length, material):
    self.length = length
    self.material = material

  def info(self):
    return f"Structural member, length = {self.length:0.2f}, material = {self.material}"



#Child class of Structural_Member without own constructor
class rod(Structural_Member):
  def axial_stress(self, force, area):
    stress = force/area
    return stress

#testing rod object

rod_1 = rod(10, "Iron")
print(f"The material of the rod_1 is {rod_1.material}")
print(f"The stress in the rod_1 is {rod_1.axial_stress(100, 20)}")
print(rod_1.info())


#Child class of Structural_Member with own constructor
class Beam(Structural_Member):
  def __init__(self, length, material, width, height):
    super().__init__(length, material)
    self.width = width
    self.height = height

  def info(self):
    print(f"{super().info()}, width ={self.width}, height = {self.height}")




#Testing of the Beam object

beam_1 = Beam(10, "Steel", 10, 5)
print(f"beam_1.length = {beam_1.length}")
print(f"beam_1.material = {beam_1.material}")
print(f"beam_1.width = {beam_1.width}")
print(f"beam_1.height = {beam_1.height}")
beam_1.info()









```

Another Interitance Example


```python
import math

#parent Class
class Structural_Member:
  def area(self):
    raise NotImplementedError("Subclass must define area()")

  def axial_stress(self, force):
    return force/self.area()

  def __str__(self):
    return self.__class__.__name__



#Child Classes
class rod(Structural_Member):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return math.pi* self.radius**2

class rectangular_beam(Structural_Member):
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def area(self):
    return self.width*self.height


#Defining object

rod1 = rod(30)
beam1 = rectangular_beam(10, 20)

members = [rod1, beam1]

for member in members:
  print(f"The axial stress in {member} is {member.axial_stress(100):0.2f} Pa")




```
