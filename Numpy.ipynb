# **Vectorized** **Operations**

```python
#Matrix Multiplication

import numpy as np

K = np.array(
    [
        [2, -1],
        [-1, 2]
    ]
)

u = np.array(
    [
        [1],
        [3]
    ]
)


#Element wise multiplication:
print("Elemet multiplication: K * u  \n", K*u, "\n"*3)

#Matrix Multiplication:
print("Matrix multiplication: K@u \n", K@u, "\n"*3) #or np.dot(K, u)




#Solving Matrix Equation with np.linalg.solve : the equation is K2. u2 = F

K2 = np.array(
    [
        [4, -1, 0],
        [-1, 4, -1],
        [0, -1, 3]
    ]
)

F = np.array(
    [
        [15],
        [10],
        [10]
    ]
)


#k2 @ u2 = F

u2 = np.linalg.solve(K2, F)

print(f"The solution of displacement is:\n\n", u2, "\n"*3)











```

# **Eigen Values And Eigen Vectors:**
a matrix A can have many (eigenvalue, eigenvector) pairs.
A square matrix of size $n \times n$ will have exactly $n$ eigenvalues, and each one will have at least one corresponding eigenvector

```python
import numpy as np


Matrix_A = np.array(
    [
        [2, -1, 0],
        [-1, 2, -1],
        [0, -1, 2]
    ]
)


lamda_A, v_A = np.linalg.eig(Matrix_A)

lamda_1 = lamda_A[0]
v_1 = v_A[:, 0]

print(f"lamda_1 * v_1: \n {lamda_1 * v_1}")
print(f"\n Matrix_A * v_1 : \n {np.dot(Matrix_A, v_1)}")


```
