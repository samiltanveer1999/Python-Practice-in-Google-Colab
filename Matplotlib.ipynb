Making a proper graph



```python
import matplotlib.pyplot as plt


time = []
y1 = []
y2 = []

for t in range (0,30):
  time.append(t)
  y1.append(t**3)
  y2.append(t*3)


plt.figure(figsize=(8,8))


plt.subplot(2,1,1)
plt.plot(time,y1, label = 'y = t^3', linewidth=2, color = 'red', marker = 'o')
plt.legend()
plt.ylabel('Position (m)')
plt.title('Comparison of two functions')
plt.minorticks_on()
plt.tick_params(axis='both', which='major', direction='in', length=10, width=1.2, top=True, right=True)
plt.tick_params(axis='both', which='minor', direction='in', length=3, width=1.2, top=True, right=True)
plt.grid(True, which='major', linewidth=0.8, linestyle='-', alpha=0.8)
plt.grid(True, which='minor', linewidth = 0.4, linestyle = ':', alpha=1)

plt.subplot(2,1,2)
plt.plot(time, y2, label = 'y = 3t', linewidth=2, linestyle ='--', color = 'green', marker = 's')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.minorticks_on()
plt.tick_params(axis='both', which='major', direction='in', length=10, width=1.2 , top=True, right=True)
plt.tick_params(axis='both', which='minor', direction='in', length=3, width=1.2, top=True, right=True)
plt.legend()
plt.grid(True, which='major', linewidth=0.8, linestyle='-', alpha=0.8)
plt.grid(True, which='minor', linewidth = 0.4, linestyle = ':', alpha=1)

plt.tight_layout()
plt.show()
```

Making the same graph with advanced function rcparams which controls the whole notebook at once.

```python
import matplotlib.pyplot as plt

# STEP 1: GLOBAL SETTINGS (rcParams)
# This ensures every subplot looks the same without repeating code.


#Global control with rcParam
plt.rcParams['font.family']='serif'
plt.rcParams['font.size']=11
plt.rcParams['axes.linewidth']=1.2
plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.top']=True
plt.rcParams['ytick.right']=True
plt.rcParams['figure.dpi']=100


#Global major tick control
plt.rcParams['xtick.major.size']=10
plt.rcParams['xtick.major.width']=1
plt.rcParams['ytick.major.size']=10
plt.rcParams['ytick.major.width']=1


#Global minor tick control
plt.rcParams['xtick.minor.size']=5
plt.rcParams['xtick.minor.width']=1
plt.rcParams['ytick.minor.size']=5
plt.rcParams['ytick.minor.width']=1





time = []
y1 = []
y2 = []

for t in range (0,30):
  time.append(t)
  y1.append(t**3)
  y2.append(t*3)


plt.figure(figsize=(8,8))

#subplot 1
plt.subplot(2,1,1)
plt.plot(time,y1, label = '$y = t^3$', linewidth=2, color = 'red', marker = 'o')
plt.ylabel('Position (m)')
plt.title('Comparison of two functions')
plt.minorticks_on()
plt.grid(True, which='major', linewidth=0.8, linestyle='-', alpha=0.8)
plt.grid(True, which='minor', linewidth = 0.4, linestyle = ':', alpha=1)
plt.annotate(
    'Yield Point',
    xy = (25, 25**3),
    xytext = (10, 25**3+5000),
    fontsize = 12,
    fontweight = 'bold',
    arrowprops = dict(facecolor = 'black', shrink=0.05, headwidth = 8, width=2,
                      # This curves the arrow to avoid the data line
                       connectionstyle="arc3, rad=-0.2")

)
plt.legend()



#subplot 2
plt.subplot(2,1,2)
plt.plot(time, y2, label = '$y = 3t$', linewidth=2, linestyle ='--', color = 'green', marker = 's')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.minorticks_on()
plt.grid(True, which='major', linewidth=0.8, linestyle='-', alpha=0.8)
plt.grid(True, which='minor', linewidth = 0.4, linestyle = ':', alpha=1)
plt.legend()

plt.tight_layout()
plt.show()
```
